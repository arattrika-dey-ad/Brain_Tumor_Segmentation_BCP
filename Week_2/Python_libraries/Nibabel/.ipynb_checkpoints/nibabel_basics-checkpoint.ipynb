## Exercise 2: Write a function plot_slices(volume, indices)

```python
# Exercise 2 Solution: plot_slices function

def plot_slices(volume, indices, titles=None, cmap='gray', figsize=None):
    """
    Plot specified slices from a 3D MRI volume.
    
    Parameters:
    -----------
    volume : numpy array
        3D MRI volume with shape (X, Y, Z)
    indices : list of int
        Slice indices to plot (Z-dimension indices)
    titles : list of str, optional
        Custom titles for each slice
    cmap : str, optional
        Colormap for display (default: 'gray')
    figsize : tuple, optional
        Figure size as (width, height)
    
    Returns:
    --------
    fig : matplotlib figure object
    axes : matplotlib axes array
    """
    
    # Validate input
    if len(volume.shape) != 3:
        raise ValueError(f"Expected 3D volume, got shape {volume.shape}")
    
    n_slices = len(indices)
    
    # Set default titles if not provided
    if titles is None:
        titles = [f'Slice {idx}' for idx in indices]
    elif len(titles) != n_slices:
        raise ValueError(f"Got {len(titles)} titles for {n_slices} slices")
    
    # Set default figsize if not provided
    if figsize is None:
        figsize = (3 * n_slices, 3)
    
    # Create figure and axes
    fig, axes = plt.subplots(1, n_slices, figsize=figsize)
    
    # Handle case when only one slice is requested
    if n_slices == 1:
        axes = [axes]
    
    # Plot each slice
    for ax, slice_idx, title in zip(axes, indices, titles):
        # Extract slice and transpose for correct orientation
        slice_data = volume[:, :, slice_idx].T
        
        # Display with grayscale colormap
        ax.imshow(slice_data, cmap=cmap)
        ax.set_title(title, fontsize=10)
        ax.axis('off')
    
    plt.tight_layout()
    return fig, axes

# ====================================================
# Test the function with your MRI data
# ====================================================

# Test 1: Plot 3 specific slices
print("Test 1: Plotting 3 specific slices")
test_indices = [5, 12, 18]  # Arbitrary slice numbers
fig1, axes1 = plot_slices(mri, test_indices, figsize=(12, 4))
plt.suptitle('Test 1: 3 Specific Slices', y=1.05, fontsize=14)
plt.show()

# Test 2: Plot 4 evenly spaced slices with custom titles
print("\nTest 2: Plotting 4 slices with custom titles")
n_slices = mri.shape[2]
even_indices = [n_slices//5, 2*n_slices//5, 3*n_slices//5, 4*n_slices//5]
even_indices = [int(i) for i in even_indices]

custom_titles = ['Frontal', 'Central', 'Posterior', 'Occipital']
fig2, axes2 = plot_slices(mri, even_indices, 
                          titles=custom_titles,
                          figsize=(12, 3))
plt.show()

# Test 3: Compare slices from both MRIs
print("\nTest 3: Comparing slices from both MRIs")
fig3, axes3 = plt.subplots(2, 3, figsize=(12, 8))

comparison_slices = [2, mri.shape[2]//2, mri.shape[2]-3]

for col, slice_idx in enumerate(comparison_slices):
    # Original MRI (time point 0)
    axes3[0, col].imshow(mri[:, :, slice_idx].T, cmap='gray')
    axes3[0, col].set_title(f'Time Point 0\nSlice {slice_idx}')
    axes3[0, col].axis('off')
    
    # Another MRI (time point 1)
    axes3[1, col].imshow(mri_another[:, :, slice_idx].T, cmap='gray')
    axes3[1, col].set_title(f'Time Point 1\nSlice {slice_idx}')
    axes3[1, col].axis('off')

plt.tight_layout()
plt.suptitle('Slice Comparison: Time Point 0 vs Time Point 1', y=1.02, fontsize=14)
plt.show()
```

## Exercise 3: Create a function to normalize all slices

```python
# Exercise 3 Solution: Normalize all slices

def normalize_volume(volume, method='global', return_stats=False):
    """
    Normalize a 3D MRI volume.
    
    Parameters:
    -----------
    volume : numpy array
        3D MRI volume
    method : str
        Normalization method:
        - 'global': Use min/max of entire volume (default)
        - 'slice': Normalize each slice independently
        - 'zscore': Zero mean, unit variance (volume-wise)
        - 'zscore_slice': Zero mean, unit variance (slice-wise)
    return_stats : bool
        If True, return normalization statistics
    
    Returns:
    --------
    normalized_volume : numpy array
        Normalized volume with same shape
    stats : dict (optional)
        Dictionary with normalization statistics
    """
    
    # Validate input
    if len(volume.shape) != 3:
        raise ValueError(f"Expected 3D volume, got shape {volume.shape}")
    
    # Initialize normalized volume
    normalized = np.zeros_like(volume, dtype=np.float32)
    stats = {}
    
    if method == 'global':
        # Global min-max normalization (like in your notebook)
        vmin, vmax = np.min(volume), np.max(volume)
        if vmax > vmin:  # Avoid division by zero
            normalized = (volume - vmin) / (vmax - vmin)
        else:
            normalized = volume.copy()
        
        stats = {'method': 'global', 'min': vmin, 'max': vmax}
    
    elif method == 'slice':
        # Slice-wise min-max normalization
        vmin, vmax = np.min(volume), np.max(volume)
        
        for z in range(volume.shape[2]):
            slice_data = volume[:, :, z]
            s_min, s_max = np.min(slice_data), np.max(slice_data)
            
            if s_max > s_min:
                normalized[:, :, z] = (slice_data - s_min) / (s_max - s_min)
            else:
                normalized[:, :, z] = slice_data
        
        stats = {'method': 'slice', 'volume_min': vmin, 'volume_max': vmax}
    
    elif method == 'zscore':
        # Volume-wise z-score normalization
        mean, std = np.mean(volume), np.std(volume)
        if std > 0:
            normalized = (volume - mean) / std
        else:
            normalized = volume.copy()
        
        stats = {'method': 'zscore', 'mean': mean, 'std': std}
    
    elif method == 'zscore_slice':
        # Slice-wise z-score normalization
        mean, std = np.mean(volume), np.std(volume)
        
        for z in range(volume.shape[2]):
            slice_data = volume[:, :, z]
            s_mean, s_std = np.mean(slice_data), np.std(slice_data)
            
            if s_std > 0:
                normalized[:, :, z] = (slice_data - s_mean) / s_std
            else:
                normalized[:, :, z] = slice_data
        
        stats = {'method': 'zscore_slice', 'volume_mean': mean, 'volume_std': std}
    
    else:
        raise ValueError(f"Unknown method: {method}. "
                         f"Choose from ['global', 'slice', 'zscore', 'zscore_slice']")
    
    if return_stats:
        return normalized, stats
    else:
        return normalized

# ====================================================
# Test the normalization function
# ====================================================

print("Testing different normalization methods:\n")

# Test each method
methods = ['global', 'slice', 'zscore', 'zscore_slice']
results = {}

for method in methods:
    print(f"\nMethod: {method}")
    
    # Normalize the volume
    normalized_vol, stats = normalize_volume(mri, method=method, return_stats=True)
    results[method] = normalized_vol
    
    # Print statistics
    print(f"  Original range: {mri.min():.2f} to {mri.max():.2f}")
    print(f"  Normalized range: {normalized_vol.min():.2f} to {normalized_vol.max():.2f}")
    print(f"  Normalized mean: {normalized_vol.mean():.4f}, std: {normalized_vol.std():.4f}")
    
    # Store stats
    for key, value in stats.items():
        if key != 'method':
            print(f"  {key}: {value:.4f}")

# ====================================================
# Visualize normalized slices
# ====================================================

# Compare different normalization methods at the same slice
slice_idx = mri.shape[2] // 2

fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Original slice
axes[0, 0].imshow(mri[:, :, slice_idx].T, cmap='gray')
axes[0, 0].set_title(f'Original\nSlice {slice_idx}')
axes[0, 0].axis('off')

# Your notebook's normalization (for comparison)
notebook_norm = (mri[:, :, slice_idx] - np.min(mri)) / (np.max(mri) - np.min(mri))
axes[0, 1].imshow(notebook_norm.T, cmap='gray')
axes[0, 1].set_title(f'Notebook Method\n(Global)')
axes[0, 1].axis('off')

# Global normalization
axes[0, 2].imshow(results['global'][:, :, slice_idx].T, cmap='gray')
axes[0, 2].set_title('Global Normalization')
axes[0, 2].axis('off')

# Slice normalization
axes[1, 0].imshow(results['slice'][:, :, slice_idx].T, cmap='gray')
axes[1, 0].set_title('Slice-wise Normalization')
axes[1, 0].axis('off')

# Z-score normalization
axes[1, 1].imshow(results['zscore'][:, :, slice_idx].T, cmap='gray')
axes[1, 1].set_title('Z-score Normalization')
axes[1, 1].axis('off')

# Z-score slice normalization
axes[1, 2].imshow(results['zscore_slice'][:, :, slice_idx].T, cmap='gray')
axes[1, 2].set_title('Z-score Slice-wise')
axes[1, 2].axis('off')

plt.tight_layout()
plt.suptitle('Comparison of Normalization Methods', y=1.02, fontsize=16)
plt.show()

# ====================================================
# Bonus: Apply normalization to "another MRI"
# ====================================================

print("\nApplying normalization to 'another MRI':")
mri_another_norm = normalize_volume(mri_another, method='global')

# Compare before/after for another MRI
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

compare_slice = mri_another.shape[2] // 3

# Original
axes[0].imshow(mri_another[:, :, compare_slice].T, cmap='gray')
axes[0].set_title(f'Original\nSlice {compare_slice}')
axes[0].axis('off')

# Normalized
axes[1].imshow(mri_another_norm[:, :, compare_slice].T, cmap='gray')
axes[1].set_title('Normalized (Global)')
axes[1].axis('off')

# Histogram of original
axes[2].hist(mri_another.flatten(), bins=50, alpha=0.7, color='blue')
axes[2].set_title('Original Histogram')
axes[2].set_xlabel('Intensity')
axes[2].set_ylabel('Frequency')

# Histogram of normalized
axes[3].hist(mri_another_norm.flatten(), bins=50, alpha=0.7, color='green')
axes[3].set_title('Normalized Histogram')
axes[3].set_xlabel('Intensity (0-1)')
axes[3].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
