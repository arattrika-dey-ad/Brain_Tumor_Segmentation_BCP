```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Assignment 1
'''
Create a numpy array of length 1000 which resembles a normal distribution with mean 0 and standard deviation 1.
Create 2 subplots both histograms with 20 bins side by side, one which has a beta (a=2, b=5) distribution, the other which has the normal distribution.
'''
```

```python
# TODO

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set random seed for reproducibility
np.random.seed(42)

# 1. Create a numpy array of length 1000 with normal distribution (mean=0, std=1)
normal_array = np.random.normal(loc=0, scale=1, size=1000)

# 2. Create a beta distribution array (a=2, b=5)
beta_array = np.random.beta(a=2, b=5, size=1000)

# 3. Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 4. Plot Beta distribution histogram
axes[0].hist(beta_array, bins=20, color='skyblue', edgecolor='black', alpha=0.7, density=True)
axes[0].set_title('Beta Distribution (a=2, b=5)', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Density')
axes[0].grid(alpha=0.3)

# Add theoretical Beta PDF for comparison
x_beta = np.linspace(0, 1, 100)
pdf_beta = stats.beta.pdf(x_beta, a=2, b=5)
axes[0].plot(x_beta, pdf_beta, 'r-', linewidth=2, label='Theoretical PDF')
axes[0].legend()

# 5. Plot Normal distribution histogram
axes[1].hist(normal_array, bins=20, color='lightgreen', edgecolor='black', alpha=0.7, density=True)
axes[1].set_title('Normal Distribution (=0, =1)', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Density')
axes[1].grid(alpha=0.3)

# Add theoretical Normal PDF for comparison
x_norm = np.linspace(-4, 4, 100)
pdf_norm = stats.norm.pdf(x_norm, loc=0, scale=1)
axes[1].plot(x_norm, pdf_norm, 'r-', linewidth=2, label='Theoretical PDF')
axes[1].legend()

# 6. Adjust layout and display
plt.suptitle('Comparison of Beta and Normal Distributions', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# 7. Print summary statistics
print("=== Normal Distribution (N=1000) ===")
print(f"Mean: {normal_array.mean():.4f}")
print(f"Standard Deviation: {normal_array.std():.4f}")
print(f"Minimum: {normal_array.min():.4f}")
print(f"Maximum: {normal_array.max():.4f}")
print()
print("=== Beta Distribution (a=2, b=5) ===")
print(f"Mean: {beta_array.mean():.4f} (Theoretical: {2/(2+5):.4f})")
print(f"Standard Deviation: {beta_array.std():.4f}")
print(f"Minimum: {beta_array.min():.4f}")
print(f"Maximum: {beta_array.max():.4f}")
```

```python
# Assignment 2
'''
Generate a NumPy array of 1000 random values drawn from a normal distribution with mean 0 and std. dev 1
Plot a histogram of the data with 20 bins.
Add vertical lines to indicate the mean of the data, the 25th percentile and the 75th percentile.
Annotate the plot with labels for the mean, 25th percentile, and 75th percentile.
Customize the visualization (light pink colour scheme, titles, labels, legend).
'''
```

```python
# TODO

import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# 1. Generate 1000 random values from normal distribution N(0, 1)
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# 2. Calculate statistics
mean_value = np.mean(normal_data)
p25 = np.percentile(normal_data, 25)  # 25th percentile
p75 = np.percentile(normal_data, 75)  # 75th percentile

# 3. Create the visualization
plt.figure(figsize=(12, 7))

# Plot histogram with light pink color scheme
hist_colors = ['#FFB6C1', '#FF69B4', '#FF1493']  # Light pink to hot pink gradient
n, bins, patches = plt.hist(normal_data, bins=20, edgecolor='#8B0A50', 
                           color=hist_colors[0], alpha=0.8, 
                           label='Normal Distribution Data')

# Gradient effect for histogram bars (optional visual enhancement)
for i, patch in enumerate(patches):
    patch.set_facecolor(hist_colors[i % len(hist_colors)])
    patch.set_alpha(0.7 + (i % 3) * 0.1)

# 4. Add vertical lines for statistics
plt.axvline(x=mean_value, color='#C71585', linestyle='-', linewidth=3, 
           label=f'Mean (={mean_value:.3f})')
plt.axvline(x=p25, color='#DB7093', linestyle='--', linewidth=2.5, 
           label=f'25th Percentile (Q1={p25:.3f})')
plt.axvline(x=p75, color='#FF69B4', linestyle='--', linewidth=2.5, 
           label=f'75th Percentile (Q3={p75:.3f})')

# 5. Add annotations with arrows
arrowprops = dict(arrowstyle='->', color='#8B008B', lw=1.5)

plt.annotate(f'Mean = {mean_value:.3f}', 
            xy=(mean_value, max(n)*0.8), 
            xytext=(mean_value+0.8, max(n)*0.85),
            arrowprops=arrowprops,
            fontsize=11, fontweight='bold', color='#8B0000',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='#FFE4E1', alpha=0.8))

plt.annotate(f'25th %ile = {p25:.3f}', 
            xy=(p25, max(n)*0.6), 
            xytext=(p25-1.2, max(n)*0.65),
            arrowprops=arrowprops,
            fontsize=10, color='#8B0000',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='#FFE4E1', alpha=0.8))

plt.annotate(f'75th %ile = {p75:.3f}', 
            xy=(p75, max(n)*0.7), 
            xytext=(p75+0.8, max(n)*0.75),
            arrowprops=arrowprops,
            fontsize=10, color='#8B0000',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='#FFE4E1', alpha=0.8))

# 6. Customize the plot
plt.title('Normal Distribution (=0, =1) with Statistical Markers', 
         fontsize=16, fontweight='bold', color='#8B0000', pad=20)
plt.xlabel('Value', fontsize=13, fontweight='bold', color='#8B0000')
plt.ylabel('Frequency', fontsize=13, fontweight='bold', color='#8B0000')

# Set pink-ish background
plt.gca().set_facecolor('#FFF0F5')
plt.grid(True, alpha=0.3, linestyle='--', color='#FFB6C1')

# Add legend with custom styling
legend = plt.legend(loc='upper right', fontsize=11, framealpha=0.9)
legend.get_frame().set_facecolor('#FFE4E1')
legend.get_frame().set_edgecolor('#FF69B4')

# Add text box with summary statistics
stats_text = (f'Sample Size: 1000\n'
              f'Standard Deviation: {np.std(normal_data):.3f}\n'
              f'Range: [{normal_data.min():.2f}, {normal_data.max():.2f}]\n'
              f'IQR (Q3-Q1): {p75-p25:.3f}')
plt.text(0.02, 0.98, stats_text, transform=plt.gca().transAxes,
         fontsize=10, verticalalignment='top',
         bbox=dict(boxstyle='round', facecolor='#FFE4E1', 
                  edgecolor='#FF69B4', alpha=0.8))

# Adjust layout
plt.tight_layout()
plt.show()

# 7. Print the calculated values
print("="*50)
print("STATISTICAL SUMMARY")
print("="*50)
print(f"Generated {len(normal_data)} values from N(0,1)")
print(f"Mean: {mean_value:.4f}")
print(f"25th Percentile (Q1): {p25:.4f}")
print(f"75th Percentile (Q3): {p75:.4f}")
print(f"Standard Deviation: {np.std(normal_data):.4f}")
print(f"Interquartile Range (IQR): {p75-p25:.4f}")
print(f"Data Range: [{normal_data.min():.4f}, {normal_data.max():.4f}]")
print("="*50)
```
