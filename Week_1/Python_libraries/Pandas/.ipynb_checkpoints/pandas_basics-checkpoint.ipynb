```python
# Assignment 1
'''
Import videogamesales.csv and perform the following actions on it
Add a column of 'global_sales' showing the total sales of all the different regions to the data frame and sort (highest first) and print the DataFrame according to it.
Display a plot of the total number of copies sold of each genre globally.
Filter out only the games containing 'Grand Theft Auto' in their name and display the following as a DataFrame:
a. their name
b. the platform they were released on
c. the year they were released in
d. the sum of sales in only Europe and Japan
Display a pie chart of the total sales of all Grand Theft Auto games combined in North America, Europe, Japan, and other sales.
'''
```

```python
# TODO

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('videogamesales.csv')

# --- Part 1: Global Sales Column ---
# Add a column of 'global_sales' showing the total sales of all regions
df['global_sales'] = df['NA_Sales'] + df['EU_Sales'] + df['JP_Sales'] + df['Other_Sales']

# Sort (highest first) and print the DataFrame
df_sorted = df.sort_values(by='global_sales', ascending=False)
print("Top 5 Games by Global Sales:")
print(df_sorted.head())

# --- Part 2: Genre Sales Plot ---
# Calculate total sales per genre
genre_sales = df.groupby('Genre')['global_sales'].sum().sort_values(ascending=False)

# Display a plot
plt.figure(figsize=(12, 6))
genre_sales.plot(kind='bar', color='skyblue')
plt.title('Total Number of Copies Sold of Each Genre Globally')
plt.xlabel('Genre')
plt.ylabel('Global Sales (Millions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('genre_sales_bar.png')
plt.show()

# --- Part 3: Grand Theft Auto Filter ---
# Filter out only games containing 'Grand Theft Auto'
# We use case=False to ignore capitalization and na=False to handle missing values safely
gta_df = df[df['Name'].str.contains('Grand Theft Auto', case=False, na=False)].copy()

# Calculate the sum of sales in only Europe and Japan
gta_df['EU_JP_Sales'] = gta_df['EU_Sales'] + gta_df['JP_Sales']

# Display the specific columns requested
print("\nGrand Theft Auto Games (Name, Platform, Year, EU+JP Sales):")
print(gta_df[['Name', 'Platform', 'Year', 'EU_JP_Sales']])

# --- Part 4: Pie Chart of GTA Sales ---
# Sum the sales for each region for all GTA games combined
gta_total_sales = gta_df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# Display the pie chart
plt.figure(figsize=(8, 8))
plt.pie(gta_total_sales, 
        labels=gta_total_sales.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Total Sales Distribution of Grand Theft Auto Games')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.savefig('gta_sales_pie.png')
plt.show()
```
